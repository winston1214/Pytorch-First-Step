```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import numpy as np
```

```python
num_data = 1000
num_epoch = 500
x = init.uniform_(torch.Tensor(num_data,1),-10,10) # 균등분포로 초기화
noise = init.normal_(torch.FloatTensor(num_data,1),std=1) # 잡음은 정규분포
y=2*x+3
y_noise = 2*(x+noise)+3
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x.numpy(),y_noise.numpy(),s=7,c="gray")
plt.axis([-12, 12, -25, 25])
plt.show()
```

```python
model = nn.Linear(1,1)
loss_func = nn.L1Loss() # MAE
```

```python
optimizer = optim.SGD(model.parameters(),lr=0.01) # 경사하강법 = stochastic gradient descent
```

```python
loss_array =[]
label = y_noise
for i in range(num_epoch):
    optimizer.zero_grad() # 최적화 함수 초기화
    output = model(x)
    
    loss = loss_func(output,label)
    loss.backward() # 손실함수 기울기
    optimizer.step() # 기울기 업뎃
    
    if i%10 == 0:
        loss_array.append(loss.data)
```

```python
import matplotlib.pyplot as plt
plt.plot(loss_array) # 손실 그래프
plt.title('loss plot')
plt.show()
```

```python
param_list = list(model.parameters())
print("Weight:",param_list[0].item(),"\nBias:  ",param_list[1].item())
```
